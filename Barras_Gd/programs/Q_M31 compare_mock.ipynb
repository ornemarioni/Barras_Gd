```python
import numpy as np

path = '/home/omarioni/Barras_GdGs/Barras_Gd/_data/Toomre/'
data = np.loadtxt(path + 'M31GD_496_toomre.dat')
M31GD_rbin   = data[:,0]
M31GD_sigmaz = data[:,1]
M31GD_sigmaR = data[:,2]
M31GD_surden = data[:,4]
M31GD_Vcirc  = data[:,5]
M31GD_Q      = data[:,6]
M31GD_kappa = M31GD_Q * 3.36 * 4.299e-6 * M31GD_surden / M31GD_sigmaR
```

```python
import numpy as np

path = '/home/omarioni/Barras_GdGs/Barras_Gd/_data/Toomre/'
data = np.loadtxt(path + 'M31GS_1024_toomre.dat')
M31GS_rbin   = data[:,0]
M31GS_sigmaz = data[:,1]
M31GS_sigmaR = data[:,2]
M31GS_surden = data[:,4]
M31GS_Vcirc  = data[:,5]
M31GS_Q      = data[:,6]
M31GS_kappa = M31GS_Q * 3.36 * 4.299e-6 * M31GS_surden / M31GS_sigmaR
```

```python
nbin = 1000
rbin = np.linspace(0,16,nbin)

QGD = np.interp(rbin, M31GD_rbin, M31GD_Q)
QGS = np.interp(rbin, M31GS_rbin, M31GS_Q)

sRGD = np.interp(rbin, M31GD_rbin, M31GD_sigmaR)
sRGS = np.interp(rbin, M31GS_rbin, M31GS_sigmaR)

SGD = np.interp(rbin, M31GD_rbin, M31GD_surden)
SGS = np.interp(rbin, M31GS_rbin, M31GS_surden)

KGD = np.interp(rbin, M31GD_rbin, M31GD_kappa)
KGS = np.interp(rbin, M31GS_rbin, M31GS_kappa)
```

```python
diff = QGD-QGS

for i in range(nbin-1):
    if np.sign(diff[i]) != np.sign(diff[i+1]):
        print('Here!')
        print (rbin[i+1]+rbin[i])/2.
```
