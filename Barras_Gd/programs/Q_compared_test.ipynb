## We compared Q of M33 and a mock galaxy

```python
import numpy as np

path = '/home/omarioni/Barras_GdGs/Barras_Gd/_data/Toomre/test/'
data = np.loadtxt(path + 'Q_test.dat')
Mock_rbin   = data[:,0]
Mock_sigmaz = data[:,1]
Mock_sigmaR = data[:,2]
Mock_surden = data[:,4]
Mock_Vcirc  = data[:,5]
Mock_rvc    = data[:,6]
Mock_Q      = data[:,7]
Mock_kappa  = data[:,8]
```

```python
import numpy as np

path = '/home/omarioni/Barras_GdGs/Barras_Gd/_data/Toomre/test/'
data = np.loadtxt(path + 'M33GD_test_toomre.dat')
M33GD_rbin   = data[:,0]
M33GD_sigmaz = data[:,1]
M33GD_sigmaR = data[:,2]
M33GD_surden = data[:,4]
M33GD_Vcirc  = data[:,5]
M33GD_Q      = data[:,6]
M33GD_kappa  = data[:,7]
```

```python
import numpy as np

path = '/home/omarioni/Barras_GdGs/Barras_Gd/_data/Toomre/test/'
data = np.loadtxt(path + 'M33GS_test_toomre.dat')
M33GS_rbin   = data[:,0]
M33GS_sigmaz = data[:,1]
M33GS_sigmaR = data[:,2]
M33GS_surden = data[:,4]
M33GS_Vcirc  = data[:,5]
M33GS_Q      = data[:,6]
M33GS_kappa  = data[:,7]
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(16, 8))#,sharey=True
#                        gridspec_kw = {'height_ratios':[3,1,3,1]}
fig.subplots_adjust(bottom=0.12, left =0.16, right = 0.95, top = 0.95)

ax[0].plot(Mock_rbin, Mock_Q,'-', lw=2.5, color='k', label='Mock')
ax[0].plot(M33GD_rbin, M33GD_Q,'-', lw=2.5, color='r', label= 'Gadget')
ax[0].plot(M33GS_rbin, M33GS_Q,'-', lw=2.5, color='b', label= 'Gasoline')
ax[0].set_ylim(0,5)
ax[0].minorticks_on()
ax[0].tick_params( labelsize=15)
ax[0].tick_params('both', length=5, width=1.2,which='minor', direction='in', right=True,top=True)
ax[0].tick_params('both', length=8, width=1.2,which='major', direction='in', right=True,top=True)
ax[0].set_ylabel('Q',fontsize=15)
ax[0].set_xlabel('R [kpc]',fontsize=15)
ax[0].set_xlim(0,15)
ax[0].legend(frameon=False, fontsize=15)

ax[1].plot(M33GD_rbin, M33GD_Q/Mock_Q,'-', lw=2.5, color='r', label= 'Gadget')
ax[1].plot(M33GS_rbin, M33GS_Q/Mock_Q,'-', lw=2.5, color='b', label= 'Gasoline')
# ax[1].set_ylim(0,5)
ax[1].minorticks_on()
ax[1].tick_params( labelsize=15)
ax[1].tick_params('both', length=5, width=1.2,which='minor', direction='in', right=True,top=True)
ax[1].tick_params('both', length=8, width=1.2,which='major', direction='in', right=True,top=True)
ax[1].set_ylabel('Qgalaxy / Qmock',fontsize=15)
ax[1].set_xlabel('R [kpc]',fontsize=15)
ax[1].set_xlim(0,15)
ax[1].legend(frameon=False, fontsize=15)

plt.show()
```
