```python
from __future__ import unicode_literals

matplotlib.rcParams['text.usetex'] = True
matplotlib.rcParams['text.latex.unicode'] = True

# cm0 = np.loadtxt('part_barra2.dat')
# tform1 = cm0[:,1]
# formt = cm0[:,3]

# cm1 = np.loadtxt('part_barra_color.dat')
# ID0 = cm1[:,0]
# ID1 = cm1[:,1]
# ID2 = cm1[:,2]
# ID3 = cm1[:,3]
# ID4 = cm1[:,4]

cm = np.loadtxt('/z/omarioni/pyprogram/newdata/tform_barra.dat')
IDs   = cm[:,0]
tform = cm[:,1]
rn_z0 = cm[:,2]
r0    = cm[:,3]

# r0_com = r0/formt

fig=plt.figure(1, figsize=(8,8))
fig.subplots_adjust(bottom=0.12, left =0.16, right = 0.95, top = 0.95)
ax=fig.add_subplot(111)
ax.plot(tform,r0,'.', markersize=0.1, color='k')
ax.set_yscale('log')
ax.set_xlim(0,14)
ax.set_ylim(0.001,1000)
# ax.spines['bottom'].set_color('w')
# ax.spines['top'].set_color('w')
# ax.spines['left'].set_color('w')
# ax.spines['right'].set_color('w')
ax.minorticks_on()
ax.tick_params( labelsize=22)#, colors='lightgray')
ax.tick_params('both', length=3, width=1.2,which='minor', direction='in', right='on',top='on')#,colors='w')
ax.tick_params('both', length=6, width=1.2,which='major', direction='in', right='on',top='on')#,colors='w')  
# ax.set_facecolor('silver')
# ax.grid(color='w',linestyle=':',which='major')
#ax.grid(color='lightgray',linestyle=':',which='minor')
ax.set_xlabel(r'$t_{form}\:[Gyr]$', fontsize=26)#, color='lightgray')
ax.set_ylabel(r'$r_{t_{form}}\:[kpc]$',fontsize=26)#, color='lightgray')
ax.axhline(y=2.5, color='r', linestyle='--')
# fig.savefig('/home/omarioni/pyprogram_4096/dist_tform.png', dpi = 100)
plt.show()
```
