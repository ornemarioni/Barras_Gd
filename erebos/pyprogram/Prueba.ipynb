```python
"""Reading the simulations"""

import matplotlib.pyplot as plt 
import numpy as np
import pynbody


s=pynbody.snapshot.load('/srv/cosmdatc/clues/B64_WM3_186592/LG/GAS_SFR/4096_Gasoline/1024/WMAP3.CLUES.HR.001024')
h = s.halos()
```

```python
#cen_x=[]
#cen_y=[]
#cen_z=[]

for i in range(1,10):
    h1 = h[i]
    cen_pot = pynbody.analysis.halo.center(h1,mode='pot',retcen=True).in_units('kpc')
    #cen_x.append(cen_pot[0])
    #cen_y.append(cen_pot[0])
    #cen_z.append(cen_pot[0])
    print 'leyendo halo', i
    xstr=(h1.star['pos'].in_units('kpc'))[:,0]
    ystr=(h1.star['pos'].in_units('kpc'))[:,1]
    zstr=(h1.star['pos'].in_units('kpc'))[:,2]
    print '-----------------'
    
    fig=plt.figure(1, figsize=(8,8))
    fig.subplots_adjust(bottom=0.12, left =0.16, right = 0.95, top = 0.95)
    
    ax1=fig.add_subplot(111)
    ax1.plot(xstr,ystr,'.', markersize=0.5, color = 'k')
    ax1.plot(cen_pot[0],cen_pot[1],'.', markersize=7, color = 'red')
    #ax.set_ylim(-1000,1000)
    #ax.set_xlim(-0.7,0)
    ax1.set_title('Nueva simulacion - Estrellas XY', fontsize=22)
    ax1.set_xlabel(u'$x\;[kpc]$', fontsize=24)
    ax1.set_ylabel(u'$y\;[kpc]$', fontsize=24)
    ax1.minorticks_on()
    ax1.tick_params( labelsize=20)
    ax1.tick_params('both', length=3, width=1.2,which='minor')
    ax1.tick_params('both', length=6, width=1.2,which='major')
    
    plt.text(x = cen_pot[0], y = cen_pot[1], s = str('%s'%i), fontsize = 20, 
            verticalalignment = 'center', horizontalalignment = 'center', color = 'g')
    
    fig=plt.figure(2, figsize=(8,8))
    fig.subplots_adjust(bottom=0.12, left =0.16, right = 0.95, top = 0.95)
    ax2=fig.add_subplot(111)
    ax2.plot(xstr,zstr,'.', markersize=0.5, color = 'k')
    ax2.plot(cen_pot[0],cen_pot[2],'.', markersize=7, color = 'red')
    #ax.set_ylim(-1000,1000)
    #ax.set_xlim(-0.7,0)
    ax2.set_title('Nueva simulacion - Estrellas XZ', fontsize=22)
    ax2.set_xlabel(u'$x\;[kpc]$', fontsize=24)
    ax2.set_ylabel(u'$z\;[kpc]$', fontsize=24)
    ax2.minorticks_on()
    ax2.tick_params( labelsize=20)
    ax2.tick_params('both', length=3, width=1.2,which='minor')
    ax2.tick_params('both', length=6, width=1.2,which='major')
    
    #if i == 9:
    #    plt.text(x = cen_pot[0], y = cen_pot[2]-250, s = str('%s'%i), fontsize = 20, 
    #                 verticalalignment = 'center', horizontalalignment = 'center', color = 'g')
    #else:
    plt.text(x = cen_pot[0], y = cen_pot[2], s = str('%s'%i), fontsize = 20, 
              verticalalignment = 'center', horizontalalignment = 'center', color = 'g')   
    
    
    fig=plt.figure(3, figsize=(8,8))
    fig.subplots_adjust(bottom=0.12, left =0.16, right = 0.95, top = 0.95)
    ax3=fig.add_subplot(111)
    ax3.plot(ystr,zstr,'.', markersize=0.5, color = 'k')
    ax3.plot(cen_pot[1],cen_pot[2],'.', markersize=7, color = 'red')
    #ax.set_ylim(-1000,1000)
    #ax.set_xlim(-0.7,0)
    ax3.set_title('Nueva simulacion - Estrellas YZ', fontsize=22)
    ax3.set_xlabel(u'$y\;[kpc]$', fontsize=24)
    ax3.set_ylabel(u'$z\;[kpc]$', fontsize=24)
    ax3.minorticks_on()
    ax3.tick_params( labelsize=20)
    ax3.tick_params('both', length=3, width=1.2,which='minor')
    ax3.tick_params('both', length=6, width=1.2,which='major')
    
    #if i == 9:
    #    plt.text(x = cen_pot[0], y = cen_pot[2]-250, s = str('%s'%i), fontsize = 20, 
    #                 verticalalignment = 'center', horizontalalignment = 'center', color = 'g')
    #else:
    plt.text(x = cen_pot[1], y = cen_pot[2], s = str('%s'%i), fontsize = 20, 
              verticalalignment = 'center', horizontalalignment = 'center', color = 'g') 
    
plt.show()
    
```
